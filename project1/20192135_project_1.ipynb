```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv

def drawDataHyperplane(x_c1, x_c2, w, color):
    drawData(x_c1, x_c2)
    drawLine(w, color)

    
def drawData(x_c1, x_c2):
    plt.plot(x_c1[:,1], x_c1[:,2], '+b')
    plt.plot(x_c2[:,1], x_c2[:,2], 'or', mfc='none')
    
def drawLine(w, color):
    x = np.arange(-5, 10, 0.1)
    y = -w[0,0]/w[0,2]-w[0,1]/w[0,2]*x
    
    plt.plot(x, y, color)

    
def Perceptron_mis_yx(y, x, w, n_epoch, eta):
    n_sample = y.shape[0]//2
    x_c1 = x[0:n_sample-1, :]
    x_c2 = x[n_sample:n_sample*2-1, :]
    yx = np.tile(y, 3)*x
    
    print('----- Perceptron_mis_yx() started -----\n')
    for e in range(1, n_epoch):
        tmp = np.tile(w, (n_sample*2, 1))
        wyx = tmp*yx
        wyx = wyx.sum(axis=1)
        #wyx = wyx.reshape(40,1)
        idx = np.where(wyx<0)
        
        yx_misclassified = yx[idx, :]
        w = w + yx_misclassified.sum(axis=1)
        
        idx = np.array(idx)
        err = idx.size
        print('err = %d' %err)
        plt.xlabel('X-Axis')
        plt.ylabel('Y-Axis')
        plt.xlim([-6, 12])      # X축의 범위: [xmin, xmax]
        plt.ylim([-5, 10])     # Y축의 범위: [ymin, ymax]
        plt.yticks([-5, 0, 5, 10])
        drawDataHyperplane(x_c1, x_c2, w, 'k')
        
        if(err == 0):
            break
            
        
    return e
       
def Perceptron_l2(y, x, w, n_epoch, eta):
    n_sample = y.shape[0]//2
    x_c1 = x[0:n_sample-1, :]
    x_c2 = x[n_sample:n_sample*2-1, :]
    
    print('----- Perceptron_l2() started -----\n')
    for e in range(1, n_epoch):
        tmp = np.tile(w, (n_sample*2, 1))
        wx = tmp*x
        wx = wx.sum(axis=1)
        wx = wx.reshape(40,1)
        
        tmp2 = np.tile(y-wx, (1,3))
        diff = tmp2*x
        
        w = w + diff.sum(axis = 0)*eta
        A = np.power(y-wx, 2)
        err1 = 1/2*A.sum(axis=0)
        
        #err2 = sum(y.*wx<0);
        idx = np.where((y*wx)<0)
        idx = np.array(idx)
        err2 = idx.size
        print('err(%d) = %f %d\n'%(e ,err1, err2))
        #plt.figure(2)
        plt.xlabel('X-Axis')
        plt.ylabel('Y-Axis')
        plt.xlim([-5, 10])      # X축의 범위: [xmin, xmax]
        plt.ylim([-5, 10])     # Y축의 범위: [ymin, ymax]
        plt.yticks([-5, 0, 5, 10])
        drawDataHyperplane(x_c1, x_c2, w, 'k')
        
        if(err2==0):
            break
        
    return e

        
def SVM(y, x, w, n_epoch, eta):
    f1 = open("x_c1.csv", "r", encoding="utf-8-sig")
    f2 = open("x_c2.csv", "r", encoding="utf-8-sig")

    data = csv.reader(f1)
    data2 = csv.reader(f2)

    A = np.ones((n_sample, 1)) #2차원
    x_c1 = np.array(list(data)) #2차원
    x_c2 = np.array(list(data2))

    x_c1 = np.hstack((A, x_c1)) #행렬 합치기 20x3
    x_c2 = np.hstack((A, x_c2))

    x_c1 = x_c1.astype(float) #string -> float
    x_c2 = x_c2.astype(float)


    #label
    y1 = np.empty((n_sample,1))
    y1.fill(-1)
    y2 = np.ones((n_sample,1))

    x = np.vstack((x_c1, x_c2))
    y = np.vstack((y1, y2))

    yx = np.tile(y, 3)*x #40x3

    n_epoch = 300
    eta = 0.001

    lamda = 0.01
    w = np.random.rand(1, 3)

    print('----- SVM() started -----\n')
    plt.xlabel('X-Axis')
    plt.ylabel('Y-Axis')
    plt.xlim([-6, 12])      # X축의 범위: [xmin, xmax]
    plt.ylim([-5, 10])     # Y축의 범위: [ymin, ymax]
    plt.yticks([-5, 0, 5, 10])

    #plt.figure(1)
    
    #drawDataHyperplane(x_c1, x_c2, w, 'k')


    for e in range (1, n_epoch):
        tmp = np.tile(w, (n_sample*2, 1))
        wyx = tmp*yx
        wyx = wyx.sum(axis=1)

        err_idx = np.where(wyx<1)
        err_idx = np.array(err_idx)
        w = w + eta*(-2*lamda*w*(n_sample*2) + yx[err_idx, :].sum(axis = 1))

        err1 = err_idx.size
        err2 = wyx[err_idx].sum(axis=1)

        #fprintf('err(%d) = %f %d, w= %f %f %f, |w|=%f, |w23|=%f\n', e, err1, err2, w, sqrt(sum(w.^2)), sqrt(sum(w(2:3).^2)));
        tmp = np.power(w,2)
        tmp = tmp.sum(axis = 1)
        tmp2 = np.power(w[0,1:3], 2)
        tmp2 = tmp2.sum(axis = 0)

        print('err(%d) = %f %e, w= %f %f %f, |w|= %f, |w23|=%f\n' %(e, err1, err2, w[0,1], w[0,1],w[0,2], np.sqrt(tmp), np.sqrt(tmp2)))

        plt.xlabel('X-Axis')
        plt.ylabel('Y-Axis')
        plt.xlim([-5, 10])      # X축의 범위: [xmin, xmax]
        plt.ylim([-5, 10])     # Y축의 범위: [ymin, ymax]
        plt.yticks([-5, 0, 5, 10])
        drawDataHyperplane(x_c1, x_c2, w, 'k')

        
n_sample = 20

f1 = open("x_c1.csv", "r", encoding="utf-8-sig")
f2 = open("x_c2.csv", "r", encoding="utf-8-sig")

data = csv.reader(f1)
data2 = csv.reader(f2)

A = np.ones((n_sample, 1)) #2차원
x_c1 = np.array(list(data)) #2차원
x_c2 = np.array(list(data2))

x_c1 = np.hstack((A, x_c1)) #행렬 합치기 20x3
x_c2 = np.hstack((A, x_c2))

x_c1 = x_c1.astype(float) #string -> float
x_c2 = x_c2.astype(float)


#label
y1 = np.empty((n_sample,1))
y1.fill(-1)
y2 = np.ones((n_sample,1))

x = np.vstack((x_c1, x_c2))
y = np.vstack((y1, y2))

w = np.random.rand(1, 3)


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim([-6, 12])      # X축의 범위: [xmin, xmax]
plt.ylim([-5, 10])     # Y축의 범위: [ymin, ymax]
plt.yticks([-5, 0, 5, 10])

plt.figure(1)
drawDataHyperplane(x_c1, x_c2, w, 'k')
plt.show()

n_epoch = 300
eta = 0.001
plt.figure(2)
ret1 = Perceptron_mis_yx(y, x, w, n_epoch, eta)
plt.show()

plt.figure(3)
ret2 = Perceptron_l2(y, x, w, n_epoch, eta)
plt.show()

print('ret1 = %d, ret2 = %d\n' %(ret1, ret2))


plt.figure(4)
SVM(y, x, w, n_epoch, eta)
plt.show()


f1.close()
f2.close()
```
